```python

#function to show top directos
def Top_Directors(n,dataF):
    #group directors names by gross
    directors = dataF[['gross','director_name']].groupby('director_name').count()
    # sort lits of directors
    directors = directors.sort_values(by='gross', ascending=True).tail(n)
    #plot graph
    directors.plot(kind='barh', figsize=(6,6))
    plt.show()

    
#function to show top Actors
def Top_Actors(n,dataF):
    #group Actor 1 names by gross
    actor_1 = dataF[['gross','actor_1_name']].groupby('actor_1_name').count()
    # sort lits of Actors 1
    actor_1=actor_1.sort_values(by='gross', ascending=True).tail(n)
    #group Actor 2 names by gross
    actor_2 = dataF[['gross','actor_2_name']].groupby('actor_2_name').count()
    # sort lits of Actors 2
    actor_2=actor_2.sort_values(by='gross', ascending=True).tail(n)
    #group Actor 3 names by gross
    actor_3 = dataF[['gross','actor_3_name']].groupby('actor_3_name').count()
     # sort lits of Actors 3
    actor_3=actor_3.sort_values(by='gross', ascending=True).tail(n)
    #create a frame to join 3 actors
    frames = [ actor_3, actor_2,actor_1]
    df_keys = pd.concat(frames, keys=['actor_3_name','actor_2_name','actor_1_name'])
    #plot the graph
    df_keys.plot(kind='barh', figsize=(6,6))
    plt.show()    
```

3. Menu Option 3 – Analyse the distribution of gross earnings
If the user selects “Analyse the distribution of gross earnings” then the
program should ask the user for a start year and then for an end year
for the analysis. Using a line graph it should then display the min,
average and max gross earnings achieved by the films for each year
between the start year and end year inclusive. The line graph should
have three lines, one for max, one for average and one for min.
Your report should contain the line graph that is outputted when the
user enters a specific start and end year.

```python
#function to compare the min, mean and max gross between two given years
def gross_earnings(year_1, year_2):   
    #filter dataset by the years given
    dataF=df[df['title_year'].between(year_1, year_2, inclusive=True)]
    #get the maximum value of each year
    dataF1=dataF[['gross','title_year']].groupby('title_year').max()
    #get the mean value of each year
    dataF2=dataF[['gross','title_year']].groupby('title_year').mean()
    #get the minimum value of each year
    dataF3=dataF[['gross','title_year']].groupby('title_year').min()
    
    #merge all information in one variable
    result = pd.concat([dataF1, dataF2, dataF3], axis=1)
    result.columns = ["Max", "Mean", "Min"]
    
    #plot the graph
    result.plot(legend=('Max', 'Mean', 'Min'))
    plt.show()
    
```

If the user selects “Genre Analysis” the program should present a listing
of all unique genres included in the dataset. The user should then be
asked to input a specific genre and your program will output the mean
IMDB score of all films within that genre.

```python
#function to get the mean IMDB score for some specific Genre
def mean_IMDB_score(Genre):
    #get any row that constain the specific Genre
    result = df['genres'].str.contains(Genre)
    #print the mean of the specific Genre
    print("Mean IMDB score for [ ", Genre," ] is [",round(df[result].imdb_score.mean(),2),"]")

```
