```python
def calculateFatalAttacks(country, df):
    boolCountry= df["Country"]==country
    boolFatal = df["Fatal"]=='Y'
    boolNonFatal = df["Fatal"]=='N'
    fatalAttacks = df[boolCountry & boolFatal]
    nonFatalAttacks = df[boolCountry & boolNonFatal]
    if (len(fatalAttacks) > 0 and len(nonFatalAttacks) > 0):
        print ("Percentage of attacks that are fatal in ",
        country,": ", (len(fatalAttacks)*100.0)/(len(fatalAttacks)
        +len(nonFatalAttacks)))
```
